```python
import sys
import numpy as np
import pandas as pd
import sklearn
import keras

print("Python: {}".format(sys.version))
print("numpy: {}".format(np.__version__))
print("pandas: {}".format(pd.__version__))
print("sklearn: {}".format(sklearn.__version__))
print("keras: {}".format(keras.__version__))
```

```python
# load dataset
names = ['n_pregnant', 'glucose_concentration', 'blood_pressuer (mm Hg)', 'skin_thickness (mm)', 'serum_insulin (mu U/ml)',
        'BMI', 'pedigree_function', 'age', 'class']

df = pd.read_csv('./data/pima-indians-diabetes.csv', names = names)
```

```python
# describe the dataset
df.describe()
```

```python
df[df['glucose_concentration'] == 0]
```

```python
# preprocess the data, mark zero values as NaN and drop
cols = ['glucose_concentration', 'blood_pressuer (mm Hg)', 'skin_thickness (mm)', 'serum_insulin (mu U/ml)', 'BMI']

for col in cols:
    df[col].replace(0, np.NaN, inplace=True)
    
df.describe()
```

```python
# drop rows with missing values
df.dropna(inplace=True)

# summarize the number of rows and cols in df
df.describe()
```

```python
# convert df to numpy array
dataset = df.values
print(dataset.shape)
```

```python
print(dataset)
```

```python
# split the dataset into input X and output y
X = dataset[:, 0:-1]
y = dataset[:, -1].astype(int)

print(X.shape)
print(y.shape)
print(y[:5])
```

```python
# normalize the data
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().fit(X)
print(scaler)
```

```python
# transform the training data
X_standardized = scaler.transform(X)

data = pd.DataFrame(X_standardized)
data.describe()
```

```python
from sklearn.model_selection import GridSearchCV, KFold
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from keras.optimizers import Adam
```

```python
# define the training model
def create_model():
    # create model
    model = Sequential()
    model.add(Dense(8, input_dim=8, kernel_initializer='normal', activation='relu'))
    model.add(Dense(4, input_dim=8, kernel_initializer='normal', activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    
    
    # compile the model
    adam = Adam(lr=0.01)
    model.compile(loss='binary_crossentropy', optimizer=adam, metrics=['accuracy'])
    
    return model
    
```

```python
# define a random seed
seed = 42
np.random.seed(seed)
```

```python
# create the model
model = KerasClassifier(build_fn=create_model, verbose=1)

# define the grid search params
batch_size = [10, 20, 40]
epochs = [10, 50, 100]

# make a dictionary of the grid search params
param_grid = dict(batch_size=batch_size, epochs=epochs)

# build and fit the GridSearchCV
grid = GridSearchCV(estimator=model, param_grid=param_grid, cv=KFold(random_state=seed), verbose=10)
grid_results = grid.fit(X_standardized, y)

# summarize the result
print("Best: {}, using {}".format(grid_results.best_score_, grid_results.best_params_))

means = grid_results.cv_results_['mean_test_score']
stds = grid_results.cv_results_['std_test_score']
params = grid_results.cv_results_['params']

for mean, std, param in zip(means, stds, params):
    print("{} ({}) with: {}".format(mean, std, param))
```
